````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from matplotlib.pylab import rcParams
rcParams['figure.figsize']  = 8,8

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

%matplotlib inline
````

````python
data = load_iris()
x = data.data
y = data.target
train_x, test_x,train_y,test_y = train_test_split(x,y,test_size=0.3,random_state=5566)
````

````python
model = KNeighborsClassifier(n_neighbors = 3)
model.fit(train_x,train_y)
predicted = model.predict(test_x)
print(predicted)
print(test_y)
````

````python
print("acc=%f" % metrics.accuracy_score(y_true=test_y, y_pred=predicted))
````

````python
from sklearn.datasets import load_boston
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsRegressor
````

````python
boston = load_boston()
boston_X = boston.data
boston_y = boston.target
# 
boston_X = preprocessing.scale(boston_X, axis=0)
train_X, test_X, train_y, test_y = train_test_split(boston_X, boston_y, test_size = 0.2, random_state=5566)
````

````python
# create KNN model
model = KNeighborsRegressor(n_neighbors=3)
# Train the model using the training sets
model.fit(train_X,train_y)
#Predict Output
predicted= model.predict(test_X)
````

````python
r2=metrics.r2_score(y_true=test_y, y_pred=predicted)
mse=metrics.mean_squared_error(y_true=test_y, y_pred=predicted)
print("r2: %f\nmse: %f"%(r2,mse))
````
