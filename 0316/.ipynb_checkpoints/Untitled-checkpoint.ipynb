```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
x = np.array([ i*np.pi/180 for i in range(60,300,4) ])
y = np.sin(x) + np.random.normal(0,0.15,len(x))
df = pd.DataFrame(np.column_stack([x,y]),columns=['x','y'])
power=15
x_string = [ 'x' ]
for i in range(2,power):
    string = "x_%d" % i
    df[string] = df[['x']]**i
    x_string.extend(["x_%d" % i])
#df.extend(['x_%d'%i for i in range(2,power+1)])


def draw(drawflag,df,power):
    plt.subplot(drawflag)
    plt.tight_layout()
    plt.plot(df[['x']],df[['y']],'.',c='b')
    x_string = [ 'x' ]
    for i in range(2,power):
        x_string.extend(["x_%d" % i])

    model = LinearRegression(normalize=True)
    model.fit(df[x_string],df[['y']])
    #print(df[x_string].head())
    
    prediction = model.predict(df[x_string])
    plt.plot(df[['x']],prediction,'.',c='r')
    mae=metrics.mean_absolute_error(prediction,y)
    #r2 = metrics.r2_score(prediction, y)   
    plt.title("Power %d"%power + "\n mae %.2f"%mae)
    rss = sum( (prediction-df[['y']]).values **2)
    ret = [rss]
    ret.extend([model.intercept_])
    if (power == 1):
        ret.extend(model.coef_) 
        return ret
    for i in range(0,power):
        print(model.coef_.shape)
        #ret.extend([model.coef_[0,i]])
        ret.extend([1]) 
    #for i in range(0,power-2):
    #    ret.extend([model.coef_[0,i]])
    print(len(ret))
    return ret

# y = 1 * x_0 + 2 * x_1 + 3
                             
col = ['rss','intercept'] + ['coef_x_%d'%i for i in range(1,16)]
ind = ['pow_%d'%i for i in range(1,16)]

perf_Linear = pd.DataFrame(index=ind, columns=col)
  
perf_Linear.iloc[0,0:1+2] = draw(231,df,1)
perf_Linear.iloc[1,0:2+2] = draw(232,df,2)
perf_Linear.iloc[8,0:9+2] = draw(233,df,9)
#perf_Linear.iloc[10] = draw(234,df,11)
#perf_Linear.iloc[12] = draw(235,df,13)
#perf_Linear.iloc[14] = draw(236,df,15)
plt.show()
#perf_Linear.iloc[0] =  np.arange(17)
perf_Linear
```
